```python
print("Program 1")
sum_positive = 0

for i in range(10):
    num = int(input("Enter a number: "))

    if num == 0:
        break

    if num < 0:
        continue

    sum_positive += num

print("Final sum:", sum_positive)


# ----------- 2 -----------
print("\nProgram 2")
num = int(input("Enter a number for multiplication table: "))

for i in range(1, 13):
    product = num * i

    if product % 3 == 0:
        continue

    if product > 50:
        break

    print(num, "x", i, "=", product)


# ----------- 3 -----------
print("\nProgram 3")
secret = 7   
attempts = 0

while True:
    guess = int(input("Guess the number (1-20): "))

    if guess < 1 or guess > 20:
        continue

    attempts += 1

    if guess == secret:
        print("Correct! Attempts:", attempts)
        break
    else:
        print("Try again!")


# ----------- 4 -----------
print("\nProgram 4")
n = int(input("Enter a number: "))
count = 0

for i in range(1, n + 1):

    if i > 50:
        break

    if i % 8 == 0:
        continue

    if i % 4 == 0:
        count += 1

print("Count:", count)


# ----------- 5 -----------
print("\nProgram 5")

for i in range(1, 6):
    for j in range(1, i + 1):

        if j == 3:
            continue

        if i * j > 6:
            break

        print(j, end=" ")

    print()
```
